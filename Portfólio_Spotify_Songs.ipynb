```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_principal = pd.read_csv('/content/spotify_top_songs_audio_features.csv')
df_principal = df_principal.drop (columns=['id','mode','time_signature','speechiness','acousticness', 'instrumentalness', 'liveness', 'valence', 'loudness'])


df_principal
```

```python
df_principal.columns = ['Nome do Artista', 'Nome da Música', 'Gravadora', 'Tom', 'Dançabilidade', 'Energia', 'Tempo', 'Duração em MS', 'Semanas no topo', 'Ouvintes']
df_principal
```

```python
df_principal['Duração'] = df_principal['Duração em MS'] / 60000
df_principal = df_principal.drop (columns='Duração em MS')
df_principal
```

```python
df_organizado = df_principal.sort_values(by="Nome do Artista")
df_principal = df_organizado
df_principal
```

```python
df_principal
```

```python
contagem_por_artista = df_principal['Nome do Artista'].value_counts().reset_index(name='Contagem')
contagem_por_artista = contagem_por_artista.set_index('index')
```

```python
top_20_mais_ouvidas = contagem_por_artista['Contagem'].head(20)
nomes_musicas = df_principal['Nome da Música'].tolist()
```

```python
plt.figure(figsize=(8, 8))
plt.pie(top_20_mais_ouvidas, labels=top_20_mais_ouvidas.index, autopct='%1.1f%%', startangle=140)
plt.title('Percentual das 20 Músicas mais ouvidas por Artista')
plt.axis('equal')
plt.show()
```
